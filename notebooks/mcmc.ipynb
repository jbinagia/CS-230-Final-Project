# Wang-Landau Free Energy

## Ising Model

### h = 0.0

```python
h = 0.0
J = 0.8
T = 1.0
N = 8

model = IsingModel(h = h, J = J)
x0 = model.init_coords(N, seed = 1234)

model.draw_config(x0)
```

```python
oprm_min = -1.0
oprm_max = 1.0
nbins = N**2

wl = WangLandauSampler(model, x0, bounds = (oprm_min, oprm_max), nbins = nbins)
```

```python
%%time

nsteps = 15000000
nwl = 1000
stride = 50000

wl.run(nsteps, nwl, stride = stride, verbose = True)
```

```python
bins = wl.bins
gn = wl.gtraj[-1]

fname = "results/mcmc/ising_free_energy_h_0.0.csv"
np.savetxt(fname, np.transpose((bins, gn)))
if os.path.exists(fname):
    arr = np.loadtxt(fname)
    bins = arr[:,0]
    gn = arr[:,1]

fig, ax = plt.subplots(1, figsize = (6, 4))

ax.plot(bins, gn, "-o", c = "k")

ax.set_xlabel("m")
ax.set_ylabel("$(F - F_0) / k_B T$")

plt.show()
```
