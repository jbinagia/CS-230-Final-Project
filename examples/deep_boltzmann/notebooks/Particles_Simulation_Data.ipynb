```python
def plot_trajectory(etraj, traj):
    fig, axes = subplots(1, 2, figsize=(10, 5))
    axes[0].plot(etraj)
    axes[0].set_xlabel('timestep')
    axes[0].set_ylabel('energy')
    dtraj = np.sqrt((traj[:, 2] - traj[:, 0])**2 + (traj[:, 3] - traj[:, 1])**2)
    axes[1].plot(dtraj)
    axes[1].set_xlabel('timestep')
    axes[1].set_ylabel('dimer distance')    
    plt.tight_layout()
    return dtraj
```

```python
class TemperatureModulator:
    def __init__(self, model, temperature_factor):
        self.model = model
        self.temperature_factor = temperature_factor
        self.dim = model.dim
    def energy(self, x):
        return self.model.energy(x) / self.temperature_factor
```

Particle permutation identity
----

```python
def distance_squared_matrix(crd1, crd2):
    n = int(np.shape(crd1)[1]/2)
    crd1_x = np.tile(np.expand_dims(crd1[:, 0::2], 2), (1, 1, n))
    crd1_y = np.tile(np.expand_dims(crd1[:, 1::2], 2), (1, 1, n))
    crd2_x = np.tile(np.expand_dims(crd2[:, 0::2], 2), (1, 1, n))
    crd2_y = np.tile(np.expand_dims(crd2[:, 1::2], 2), (1, 1, n))
    Dx = crd1_x - np.transpose(crd2_x, axes=(0, 2, 1))
    Dy = crd1_y - np.transpose(crd2_y, axes=(0, 2, 1))
    D2 = Dx**2 + Dy**2
    return D2
```

```python
def hungarian(X):
    # reference frame
    x_ref = 1.05*model.init_positions(1.5)
    # solve minimal assignment problem
    solvent_assignment_costs = distance_squared_matrix(np.tile(x_ref[:, 4:], (X.shape[0], 1)), X[:, 4:])    
    # assign to new traj
    X_hungarian = []
    for t in range(X.shape[0]):
        _, col_assignment = scipy.optimize.linear_sum_assignment(solvent_assignment_costs[t])
        col_assignment = np.vstack([2*col_assignment, 2*col_assignment+1]).T.flatten()
        X_hungarian.append(np.concatenate([X[t, :4], X[t, 4:][col_assignment]]))
    X_hungarian = np.array(X_hungarian)
    
    return X_hungarian
```

```python
from deep_boltzmann.sampling.permutation import HungarianMapper
mapper = HungarianMapper(1.05*model.init_positions(1.5), dim=2, identical_particles=np.arange(2, model.nparticles))
```

```python
traj_closed_train_hungarian = mapper.map(sampler_closed.traj)
traj_open_train_hungarian = mapper.map(sampler_open.traj)
traj_closed_test_hungarian = mapper.map(sampler_closed2.traj)
traj_open_test_hungarian = mapper.map(sampler_open2.traj)
```
